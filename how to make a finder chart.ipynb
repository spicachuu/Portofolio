```python
import matplotlib.pyplot as plt
from astroplan import FixedTarget
from astroplan.plots import plot_finder_image
import astropy.units as u

# Membuat target objek langit
target = FixedTarget.from_name("M 64")

# Plot finder chart
fig, ax = plt.subplots(figsize=(8, 8))
ax, hdu = plot_finder_image(target, ax=ax, fov_radius=5*u.arcmin, reticle=True)

# Menampilkan finder chart
plt.show()
```
