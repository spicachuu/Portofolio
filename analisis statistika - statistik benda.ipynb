4. Distribusi Binomial [10 poin]
Mahasiswa SAP yang memilih KK Keplanetan sebanyak 60% sedangkan sisanya memilih KK Atmosfer.
a. Jika kita memilih secara acak 7 mahasiswa SAP, berapa peluang mendapatkan 3 mahasiswa KK Keplanetan dan 4 mahasiswa KK Atmosfer?
b. Buatlah plot probability mass function untuk kasus di atas, probabilitas mendapati 0, 1, 2, ..., hingga 7 mahasiswa memilih KK Keplanetan!

```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt

n, p = 10, 0.6
x = np.arange(0,11)
print("n :", 100)
print("p: ", 0.6)
print("k: ", x)
mean, var, skew, kurt = binom.stats(n, p, moments='mvsk')
print(f"Mean: {mean}\nVar: {var}\nSkew: {skew}\nKurt: {kurt}")

fig, ax = plt.subplots(1, 1)
ax.plot(x, binom.pmf(x, n, p), 'bo', ms=8, label='binomial pmf (n=10, p=0.6)')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)
plt.legend()
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, binom.pmf(x, n, p), 'bo', ms=8, label='binomial pmf (n=7, p=0.6)')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='b', lw=3, alpha=1.0)

ax.plot(x, binom.cdf(x, n, p), 'ro', ms=8, label='binomial cdf (n=7, p=0.6)')
ax.vlines(x, 0, binom.cdf(x, n, p), colors='r', lw=6, alpha=0.3)
plt.legend()
```

5. Distribusi Poisson [10 poin]
Eta Aquarids adalah hujan meteor yang berkaitan dengan komet Halley dan dapat
teramati antara 29 April hingga 28 Mei. Saat puncaknya pada tanggal 5 Mei jumlah
meteor yang terlihat rata-rata mencapai 55 per jam. Andaikan kita mengamati ketika
puncaknya ini, maka
a. Berapa ekspektasi jumlah meteor yang dapat teramati dalam waktu 5 menit?
b. Buatlah plot probability mass function untuk jumlah meteor yang teramati dalam
selang waktu 5 menit; dari 0 hingga 10 meteor.

```python
55 per jam
ekspektasi selama 1 menit = 4,5 meteor
```

```python
import math
4.5**4.5/math.factorial(bool(4.5))*np.exp(-4.5)
```

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt

mu = 4.5
mean, var, skew, kurt = poisson.stats(mu, moments='mvsk')
print("mean : ", mean)
print("var  : ", var)
print("skew : ", skew)
print("kurt : ", kurt)

x = np.arange(0, 11)
print("k    : ", x)
fig, ax = plt.subplots(1, 1)
ax.plot(x, poisson.pmf(x, mu), 'bo', ms=8, label='poisson pmf')
ax.vlines(x, 0, poisson.pmf(x, mu), colors='b', lw=5, alpha=0.5)
plt.legend()
```

6. Distribusi Normal [10 poin]
Curah hujan di Bandar Lampung pada bulan Maret mengikuti distribusi normal dengan
rata-rata 300 mm/bulan dan standar deviasi 50 mm/bulan.
a. Berapa peluang kita mendapatkan curah hujan lebih dari 375 mm/bulan?
b. Buatlah plot probability density function untuk curah hujan di Bandar Lampung
pada bulan Maret!

```python
(1/50*np.sqrt(2*3.14))*np.exp(-0.5*((375-300)/50)**2)
```

```python
mu = 300
sigma = 50
Nsample = 10000

s = np.random.normal(mu, sigma, Nsample) # normal/gaussian dist. random number generator
print(s)

nbins=51

from scipy.special import erf # for cdf plot

count, bins, ignored = plt.hist(s, nbins)
width = bins[1]-bins[0]
print(f"Count [{len(count)}]: ", count)
print("bins: ", bins)
print("width: ", width)
plt.plot(bins, Nsample*width * 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')
plt.show()

count, bins, ignored = plt.hist(s, nbins, cumulative=True)
plt.plot(bins, Nsample * 0.5*(1 + erf((bins-mu)/(np.sqrt(2)*sigma))), linewidth=2, color='r')
plt.show()

#####
## density True -> normalize -> pdf
count, bins, ignored = plt.hist(s, nbins, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')
plt.show()

count, bins, ignored = plt.hist(s, nbins, density=True, cumulative=True)
plt.plot(bins, 0.5*(1 + erf((bins-mu)/(np.sqrt(2)*sigma))), linewidth=2, color='r')
plt.show()
```
