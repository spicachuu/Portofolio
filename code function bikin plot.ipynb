```python
#import the libraries we used
from astropy.io import fits
from astropy.stats import sigma_clipped_stats
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
plt.style.use('seaborn-white')
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.family'] = 'STIXGeneral'
```

```python
def ngeplot(x,y,z):
    file = fits.open(x)
    filebaru = file[0].data
    fig,ax = plt.subplots(figsize=(6,4))
    cp = plt.imshow(filebaru, cmap='yarg', norm=z)
    cbar = plt.colorbar(cp)
    ax.set_xlabel('X (piksel)')
    ax.set_ylabel('Y (piksel)')
    cbar.ax.set_ylabel('ADU')
    plt.savefig(y, dpi=300, bbox_inches='tight',pad_inches=0.1)
```
